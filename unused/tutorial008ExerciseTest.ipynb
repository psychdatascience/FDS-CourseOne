## Basic relationship plots exercise test

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mdff = pd.read_csv("datasets/008ExerciseFile.csv")
display(mdff)
```

```python
sns.relplot(data=mdff, x='rel_time', y='m_level')
```

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level')
```

```python
sns.residplot(data=mdff, x='rel_time', y='m_level')
```

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=3)
```

```python
g = sns.residplot(data=mdff, x='rel_time', y='m_level', order=3)
```

```python
myfit = np.polyfit(x=mdff['rel_time'], y=mdff['m_level'], deg=3)
print(myfit)
```

```python
fitvals = np.polyval(myfit, mdff['rel_time'])
```

```python
myres = mdff['m_level'] - fitvals
```

```python
sns.displot(myres, kind="kde")
```

```python
np.std(myres)
```
